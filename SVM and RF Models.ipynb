```python
Y_cols = 'Cooling Devices'
y = df[Y_cols].values   #Should change with dataframe (df5 or df)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=30)
y_train
```

```python
y_train = y_train.ravel()
y_train
```

```python
y_test = y_test.ravel()
y_test
```

```python
from sklearn.svm import SVR
SVReg = SVR(kernel = 'rbf', C = 700, epsilon = 0.003, gamma = 0.0001) # C = 700, epsilon = 0.003, gamma = 0.0001)  C = 2000, epsilon = 0.0001, gamma = 0.0001)  C = 20, epsilon = 0.005, gamma = 0.01) C = 10, epsilon = 0.001, gamma = 0.001) , C = 1, epsilon = 0.69, gamma = 1) #regr = SVR(kernel = 'rbf', C = 10, epsilon = 0.03, gamma = 7.75),(C = 1, epsilon = 0.69, gamma = 1)
SVReg.fit(x_train,y_train)
```

```python
Predicted_train = SVReg.predict(x_train)
Predicted_train
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
print(r2_score(y_train,Predicted_train)*100)
print(mean_squared_error(y_train,Predicted_train))
print(mean_absolute_error(y_train,Predicted_train))
print(mean_absolute_percentage_error(y_train,Predicted_train))
rmse = np.sqrt(mean_squared_error(y_train,Predicted_train))
print(rmse)
cvrmse = (np.sqrt(mean_squared_error(y_train, Predicted_train)))/(np.mean(y_train))
print(cvrmse)
# CVRSME = 100 * ((sum((test_labels - predictions)**2) / (pd.Series(test_labels).count()-1))**(0.5)) / np.mean(test_labels)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RFReg = RandomForestRegressor(n_estimators = 120, max_depth = 30, n_jobs = -1, bootstrap = True, oob_score = True, 
                              random_state = 40)
x_train2, x_test2, y_train2, y_test2 = train_test_split( x1, y, test_size=0.2, random_state=30, shuffle = "False")
y_train2 = y_train2.ravel()
y_test2 = y_test2.ravel()
RFReg.fit(x_train2,y_train2)
predicted_train2 =  RFReg.predict(x_train2)
print(r2_score(y_train2,predicted_train2)*100)
print(mean_squared_error(y_train2,predicted_train2))
print(mean_absolute_error(y_train2,predicted_train2))
print(mean_absolute_percentage_error(y_train2,predicted_train2))
rmse = np.sqrt(mean_squared_error(y_train2,predicted_train2))
print(rmse)
cvrmse = (np.sqrt(mean_squared_error(y_train2, predicted_train2)))/(np.mean(y_train2))
print(cvrmse)
```

```python
plt.figure(figsize = (16,4))
plt.plot(y_train2, color = 'b', label = "Original Data")
plt.plot(predicted_train2, color = 'y', label = "RFReg Predicted Data")
plt.legend(loc = 'best')
plt.xlabel('Time')
plt.ylabel('Load in MW')
```

```python
pred2 =  RFReg.predict(x_test2)
print(r2_score(y_test2,pred2)*100)
print(mean_squared_error(y_test2,pred2))
print(mean_absolute_error(y_test2,pred2))
print(mean_absolute_percentage_error(y_test2,pred2))
rmse = np.sqrt(mean_squared_error(y_test2,pred2))
print(rmse)
cvrmse = (np.sqrt(mean_squared_error(y_test2, pred2)))/(np.mean(y_test2))
print(cvrmse)
```
